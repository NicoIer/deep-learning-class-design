```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
LeNetMind = pd.read_csv("./results/mindspore/LeNetMindspore-result.csv")
LeNetProMind = pd.read_csv("./results/mindspore/LeNetProMindspore-result.csv")
```

```python
sns.heatmap(np.array([[1,0],[0,1]]))
plt.title("confusion matrix")
plt.savefig("./fake8.png")
```

```python
# 设置线条的样式
sns.set_style("whitegrid")
# 设置绘图的风格
# sns.set_context("poster")
# 设置绘图的尺寸
# plt.figure(figsize=(10, 5))
# 设置线条的宽度
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 1})
plt.plot(LeNetMind['train_score'],label="lenet train score")
plt.plot(LeNetProMind['train_score'],label="lenetpro train score")
plt.legend()
plt.title("train score")
plt.savefig("./results/mindspore/train_score.png")
```

```python
# 设置线条的样式
sns.set_style("whitegrid")
# 设置绘图的风格
# sns.set_context("poster")
# 设置绘图的尺寸
# plt.figure(figsize=(10, 5))
# 设置线条的宽度
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 1})
plt.plot(LeNetMind['test_score'],label="lenet test score")
plt.plot(LeNetProMind['test_score'],label="lenetpro test score")
plt.legend()
plt.title("test score")
plt.savefig("./results/mindspore/test score.png")
```

```python
# 设置线条的样式
sns.set_style("whitegrid")
# 设置绘图的风格
# sns.set_context("poster")
# 设置绘图的尺寸
# plt.figure(figsize=(10, 5))
# 设置线条的宽度
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 1})
plt.plot(LeNetMind['epoch_loss'],label="lenet loss")
plt.plot(LeNetProMind['epoch_loss'],label="lenetpro loss")
plt.legend()
plt.title("loss")
plt.savefig("./results/mindspore/loss.png")
```
